```python
dates=jhop_confirmed.columns[4:]

jh1=jhop_confirmed.melt(id_vars=['Country/Region','Lat','Long'], value_vars=dates,var_name='Date',value_name='Confirmed')

jh2=jhop_recovered.melt(id_vars=['Country/Region','Lat','Long'],value_vars=dates,var_name='Date',value_name='Recovered')

jh3=jhop_death.melt(id_vars=['Country/Region','Lat','Long'],value_vars=dates,var_name='Date',value_name='Deaths')
```
